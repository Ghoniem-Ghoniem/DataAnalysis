```python
import numpy as np
import pandas as pd
OriginalDF = pd.read_excel("Budget.xlsx",'01-Budget')
df = OriginalDF[["Year","Revenue"]]
df = df.iloc[:36]
df
```

```python
groupeddf = df.groupby("Year").sum()
groupeddf
```

```python
groupeddf = df.groupby("Year").median()
groupeddf
```

```python
groupeddf = df.groupby("Year").count()
groupeddf
```

```python
table = pd.pivot_table(data=df,index=['Year'],aggfunc='mean')
table
```

```python
table = pd.pivot_table(data=df,index=['Year'],aggfunc='max')
table
```

```python
table = pd.pivot_table(data=df,index=['Year'],aggfunc='count')
table
```

```python
def GetAbsoluteMonth(MonthString):
    #return MonthString[:-2]
    return MonthString.split()[0]
GetAbsoluteMonth('أغسطس 24')
```

```python
NewDF = OriginalDF[['Year','Month','Revenue']]
NewDF= NewDF.iloc[:36]
```

```python
NewDF['AbsoluteMonth'] =NewDF['Month'].apply(GetAbsoluteMonth)
```

```python
table = pd.pivot_table(data=NewDF[['Year','AbsoluteMonth','Revenue']],index=['AbsoluteMonth'],columns='Year',aggfunc='mean')
table
```
