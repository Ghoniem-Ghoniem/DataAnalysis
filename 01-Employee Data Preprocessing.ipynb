```python
import pandas as pd
import heapq
excelfilepath = "Employees.xlsx"
df = pd.read_excel(excelfilepath,'Employees')
newdf = pd.read_excel(excelfilepath,'NewEmployees')

from sklearn.cluster import KMeans
X = df[['Salary']]
y = df[['Role']]
#X= df.Year.values
#y= df.TotalCost.values
df.describe() #Will display the numeric analytics only
```

```python
df[df.Salary.isnull()] #Validate null alues in salary, that should not be accepted
```

```python
df[df.EmpID.duplicated()] #Validate null alues in Employees, that should not be accepted
```

```python
df
```

```python
df = df.drop_duplicates(keep='last')#Drop duplicates and keep the last one
df
```

```python
df.Salary.fillna(df.Salary.mean(),inplace=True) # filling the empty salaries with the average to resolve empty data
df
```

```python
print (heapq.nlargest(5,list(df.Salary))) # Getting the most largest 5 numbers in order to validae with upper salary limits
print (heapq.nsmallest(5,list(df.Salary)))# Getting the most smallest 5 numbers in order to validae with lower salary limits
```

```python
df[df.Salary<df.Salary.mean()*.25] #Validate All Employees , that Below the quarter of average
```

```python
#Grouping datarframes
groupeddf = df.groupby('Role')
print("Group per role count\n",groupeddf.count()) # get the group by count
print("Group per Sum\n",groupeddf.sum())# get the group by sum
print("Group per Mean\n",groupeddf.mean()) # get the group by avg
groupeddf.describe() #describe the groups
```

# Sampling

```python
print("Sampling by Count",df.sample(n=5)) #get sample by count
print ("Sampling by %",df.sample(frac =.10)) #get sample by count
```

```python
#merging 2 dataframes
frames =[df,newdf]
allrecords = pd.concat(frames)
allrecords
```

```python
#Adding new columns 
allrecords['SalaryWithTax'] = allrecords['Salary']*1.22
allrecords['NewIndex'] = allrecords.index * 10
allrecords
```

```python
#Removing Column
del allrecords['NewIndex']
```

```python
#Pivot Table, with aggregation
pd.pivot_table(allrecords,values=['Salary'],columns=['Role'],aggfunc='sum')
```

```python
#Saving dataframe to external entity
import os
FilePath = os.getcwd()+'\\ConcatenatedStudents.xlsx'
writer = pd.ExcelWriter(FilePath)
allrecords.to_excel(writer,sheet_name ='ConcatenatedSheet')
writer.save()
FilePath
```
