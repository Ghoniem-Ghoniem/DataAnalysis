```python
df = df[["Year","Revenue"]]
```

```python
#AGGREGATE FUNCTIONS for describtive analysis
print (df.describe(percentiles=[.1,.2,.3,.4,.6,.7,.8,.9]))
print ("Sum of dataset is",df.sum().values)
print ("Min of dataset is",df.min().values)
print ("Max of dataset is",df.max().values)
print ("Mean of dataset is",df.mean().values)
print ("Harmonic  mean of dataset is",st.hmean(df.Revenue))
print ("Geometric mean of dataset is",st.gmean(df.Revenue))
print ("Interquartile range of dataset is",st.iqr(df.Revenue))
print ("Median of dataset is",df.median().values)
print ("First quartile of dataset is",df.quantile(.25).values) # means the quartile
print ("Second quartile of dataset is",df.quantile(.5).values) # means the quartile
print ("Third quartile of dataset is",df.quantile(.75).values) # means the quartile
print (" Mode of dataset is",df.mode().values) # means the quartile
```

```python
#AGGREGATE FUNCTIONS for Despiration measurements
print (" Standard deviation of dataset is",df.Revenue.std())
print (" Variance of dataset is",df.Revenue.var())
```
